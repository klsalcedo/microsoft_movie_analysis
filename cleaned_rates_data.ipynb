```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#read in rotten tomatoes rating file as a pandas dataframeshape(54432,8)
rtRev_df = pd.read_csv('zippedData/rt.reviews.tsv', delimiter='\t', encoding='latin1')
rtRev_df
```

```python
#created new df without all the NaN values in rating column, dropped 13517 rows, shape(40915,8)
cleaned_df = (rtRev_df[rtRev_df['rating'].notna()]).copy()
cleaned_df.head()
```

```python
#created a new df where all the rows have a letter as the value in the rating column, shape(6468,8)
df_letters = rtRev_df[rtRev_df['rating'].isin(['A+','A','A-','B+','B','B-','C+','C','C-','D+','D','D-','F+','F','F-','N','T','R'])].copy()
df_letters.head()
```

```python
#used the letters_df to rid the letters from the no-Nan cleaned_df, dropped 6468 rows, shape(34447,8)
cleaned_df.drop(cleaned_df[cleaned_df['rating'].isin(df_letters['rating'])].index, inplace=True)
```

```python
cleaned_df['rating'].unique()
```

```python
#get list of string data values from rating column
values = []
for rating_value in cleaned_df['rating']:
    values.append(rating_value)

#partition and convert string values to float, interpreted the 1-5 as 1.5
to_num = []
for item in values:
    if item == '3 1/2':
        to_num.append(3.5)
    elif '/' in item:
        x = item.partition('/')
        to_num.append(float(x[0]))
    elif item == '1-5':
        to_num.append(1.5)
    else:
        to_num.append(float(item))

#rating should be 0-5 so half any ratings that are above 5, overall 3044 ratings over 5
five_rating = []
counter = 0
for element in to_num:
    if element > 5:
        counter += 1
        y = element/2
        five_rating.append(y)
    else:
        five_rating.append(element)

#add new column 'cleaned_rating' w cleaned ratings and print counter to ascertain number of rating above 5
print(counter)
cleaned_df['cleaned_rating'] = five_rating
cleaned_df.head()
```

```python
#df with movie info, use to obtain genres, shape(1560,12)
rt_df = pd.read_csv('zippedData/rt.movie_info.tsv', delimiter='\t') 
rt_df.head()
```

```python
#merged the cleaned df w the rotten tomatoes movie info df on the id column to get genres, shape(34447,20)
w_genres_df = cleaned_df.merge(rt_df, how='left', on='id')
w_genres_df.head()
```

```python
#dropped columns that seemed superfluous, shape = (40868,14)
w_genres_df.drop(labels=['critic','publisher','top_critic','rating_y','dvd_date','studio'],axis=1, inplace=True)
w_genres_df.head()
```

```python
#made a copy of w_genres_df and dropped the rows with NaN values in genre column, dropped 59 rows, shape(34388,14)
filled_genres = w_genres_df.dropna(subset=['genre']).copy()
filled_genres.head()
```

```python
#separate the grouped genres in the genre column into individual genres
separated = []
word = ''
for genre in filled_genres['genre']:
    counter = genre.count('|') + 1
    word = genre
    while counter > 0:
        part = word.partition('|')
        separated.append(part[0])
        word = part[2]
        counter -= 1

```

```python
#get a list of the unique genres 
distinct = set(separated)
distinct
```

```python
#get boolean values for each movie for each distinct genre, will be in a list of lists
genre_by_row = []
for item in distinct:
    distinct_values = []
    for genre in filled_genres['genre']:
        if item in genre:
            distinct_values.append(True)
        else:
            distinct_values.append(False)
    genre_by_row.append(distinct_values)
        
```
